## 1. 넘파이 (Numpy)

### (1) 넘파이 (Numpy) 개요
- 파이썬에서 선형대수 기반의 프로그램을 쉽게 만들 수 있도록 지원하는 대표적인 패키지

### (2) 넘파이 배열 생성
- numpy의 기반 데이터 타입은 다차원 배열(ndarray)
- 다차원 배열이란 같은 자료형의 데이터를 담을 수 있는 하나의 그릇

#### 넘파이 임포트

```python
import numpy as np
```

#### 넘파이 배열 형태

```python
a1 = np.array([1, 2, 3])
print(a1)
print('array1 type : ', type(a1))
print('array1 shape : ', a1.shape)   # 1차원 배열에 3개의 데이터

a2 = np.array([[1, 2, 3], [4, 5, 6]])
print(a2)
print('array2 type : ', type(a2))
print('array2 shape : ', a2.shape)   # 2차원 배열에 2개의 행과 3개의 컬럼 데이터

a3 = np.array([[1, 2, 3]])
print(a3)
print('array3 type : ', type(a3))
print('array3 shape : ', a3.shape)   # 2차원 배열에 1개의 행과 3개읠 컬럼 데이터
```

#### 넘파이 배열 차원

```python
print('array1 dim :', a1.ndim)
print('array2 dim :', a2.ndim)
print('array3 dim :', a3.ndim)
```

#### 넘파이 arange 함수 기본

```python
a = np.arange(20)   # np.arange(x): 0부터 (x-1)까지의 값을 순차적 배열의 데이터 값으로 생성
print(a)
```

#### 넘파이 arange 함수 응용

```python
a = np.arange(1, 20, 3)   # np.arange(x, y, z): x값부터 (y-1)값까지 z값 만큼 증가하는 배열 생성
print(a)
```

### (3) 넘파이 배열 초기화

#### 넘파이 zeros, ones 함수

```python
zero_a = np.zeros((2,5))
one_a = np.ones((3,4))

print(zero_a)
print(one_a)
```

#### 넘파이 zeros_like 함수

```python
zero_b = np.zeros_like(one_a)   # 이미 생성한 배열의 구조와 동일한 구조를 유지한 상태에서 0으로 채워진 배열로 초기화하여 생성하는 방법
print(zero_b)
```

#### 넘파이 ones_like 함수

```python
one_b = np.ones_like(zero_a)
print(one_b)
```

#### 넘파이 full, random 함수

```python
full_a = np.full((4, 3),9)
random_a = np.random.random((3, 4))

print(full_a)
print(random_a)
```

#### 넘파이 eye 함수

```python
eye_a = np.eye(4)   # 주대각선의 원소가 모두 1이고, 나머지는 모두 0인 대각행렬 형태의 배열 생성
print(eye_a)
```

### (4) 넘파이 배열 크기 변형
- reshape: 배열의 차원과 크기를 변경
- 변환하고자 하는 크기를 함수 인자로 입력

#### 넘파이 reshape 함수 기본

```python
array1 = np.arange(12)
print('array1:\n', array1)

array2 = array1.reshape(3,4)
print('array2:\n', array2)

array3 = array1.reshape(3,4,order='F')   # order='F': reshape 함수의 옵션으로 선언하면 값을 열부터 채워 넣게 되고, 옵션을 입력하지 않으면 값을 행부터 채워 넣음
print('array3:\n', array3)
```

#### 넘파이 reshape 함수 응용

```python
array1 = np.arange(10)
array2 = array1.reshape(-1, 5)   # 인자로 -1 입력하는 경우가 일반적 / -1 입력하면 원래 배열과 호환되는 크기로 자동 변환

print(array2)
print('array2 dim : ', array2.ndim)
print('array2 shape : ', array2.shape)
```

#### 넘파이 reshape 함수 다차원 변형

```python
array1 = np.arange(12)
array2 = array1.reshape(2,3,2, order='F')

print(array2)
print('array2 dim : ', array2.ndim)
print('array2 shape : ', array2.shape)
```

#### 넘파이 flatten 함수 1차원 변형

```python
array3 = array2.flatten()   # 현재까지 구성한 다차원 배열을 1차원으로 만듦
print(array3)
```

### (5) 넘파이 배열 데이터 추출하기

#### 넘파이 1차원 배열 단일값 추출

```python
array1 = np.arange(1, 10)
print('array1 : ', array1)

value1 = array1[2]
value2 = array1[-2]

print(value1)
print(value2)
```

#### 넘파이 다차원 배열 단일값 추출

```python
array1 = np.arange(1, 10)
array2 = array1.reshape(3, 3)
print('array2 :\n', array2)

value1 = array2[0,0]
value2 = array2[-1, -1]
value3 = array2[-1, -2]

print(value1)
print(value2)
print(value3)
```

#### 넘파이 1차원 배열 슬라이싱 추출

```python
array1 = np.arange(1, 10)
array2 = array1[0:3]
array3 = array1[:3]
array4 = array1[3:]

print(type(array2))
print(array2)
print(array3)
print(array4)
```

#### 넘파이 다차원 배열 슬라이싱 추출

```python
array1 = np.arange(1, 10)
array2 = array1.reshape(3, 3)
print('array2 : \n', array2)

print('array2[0:2, 0:2]\n', array2[0:2, 0:2])
print('array2[1:3, 0:3]\n', array2[1:3, 0:3])
print('array2[:2, 1:]\n', array2[:2, 1:])
print('array2[:2, 0]\n', array2[:2, 0])
```

### (6) 넘파이 선형대수 기본연산

#### 넘파이 행렬 내적 계산
- 행렬 내적은 행렬 곱을 의미

```python
array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])
array2 = np.array([[7, 8],
                   [9, 10],
                   [11, 12]])

dot_array = np.dot(array1, array2)
print(dot_array)
```

#### 넘파이 전치 행렬 계산
- 전치 행렬: 원 행렬에서 행과 열 위치를 교환한 원소로 구성한 행렬

```python
array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])
transpose_array = np.transpose(array1)

print(transpose_array)
```

### (7) 넘파이 기술 통계

#### 배열 데이터 개수 계산

```python
x = np.array([18,   5,  10,  23,  19,  -8,  10,   0,   0,   5,   2,  15,   8,
              2,   5,   4,  15,  -1,   4,  -7, -24,   7,   9,  -6,  23, -13])

len(x)
```

#### 평균, 분산, 표준편차 계산

```python
print(np.mean(x))
print(np.var(x))
print(np.std(x))
```

#### 최대값, 최소값, 중앙값 계산

```python
print(np.max(x))
print(np.min(x))
print(np.median(x))
```

#### 사분위수 계산
- y가 25일 때, 1사분위수
- y가 50일 때, 2사분위수
- y가 75일 때, 3사분위수

```python
print(np.percentile(x, 25))
print(np.percentile(x, 50))
print(np.percentile(x, 75))
```
