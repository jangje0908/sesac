```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=10, size=(100, )) # 평균(loc)이 50이고 표준편차(scale)가 10인 정규 분포를 따르는 난수 생성

fig, ax = plt.subplots(figsize=(7, 7))
ax.boxplot(math_scores)
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=10, size=(100, ))

fig, ax = plt.subplots(figsize=(7, 7))
ax.boxplot(math_scores, notch=True)   # notch=True: 상자 그림의 중간 부분에 홈(노치) 생성
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=10, size=(100, ))

fig, ax = plt.subplots(figsize=(7, 7))
ax.boxplot(math_scores, notch=True, whis=2)   # whis=2: whis는 "whisker"를 나타내며, 상자 그림에서 수염(whisker)의 길이를 조절하는 매개변수
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=10, size=(100, ))

fig, ax = plt.subplots(figsize=(7, 7))
ax.boxplot(math_scores, notch=True, whis=1, sym='bx')   # sym='bx: 이상치를 파란색 십자로 표시
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=10, size=(100, ))

fig, ax = plt.subplots(figsize=(7, 7))
ax.boxplot(math_scores, notch=True, showfliers=False)   # showfliers=False: 이상치(outliers)를 상자 그림에서 표시하지 않음
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=10, size=(100, ))

fig, ax = plt.subplots(figsize=(7, 7))
ax.boxplot(math_scores, notch=True, showfliers=False, vert=False)   # vert=False: 상자 그림의 방향을 수평으로 변경
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

medianprops = {'linewidth': 2, 'color': 'k'}
boxprops = {'linestyle': '--', 'color': 'k', 'alpha': 0.7}
whiskerprops = {'linestyle': '--', 'color': 'b', 'alpha': 0.8}

ax.boxplot(math_scores, medianprops=medianprops, boxprops=boxprops, whiskerprops=whiskerprops, capprops=whiskerprops)
```

```python
n_student - 100
math_scores = np.random.normal(loc=50, scale=15, size=(100, 1))
chem_scores = np.random.normal(loc=70, scale=10, size=(n_student, 1))
phy_scores = np.random.normal(loc=30, scale=12, size=(n_student, 1))
pro_scores = np.random.normal(loc=80, scale=5, size=(n_student, 1))

data = np.hstack((math_scores, chem_scores, phy_scores, pro_scores))

medianprops = {'linewidth': 1.5, 'color': 'r'}
boxprops = {'linewidth': 1.5, 'color': 'k', 'alpha': 0.7}
whiskerprops = {'linestyle': '--', 'color': 'b', 'alpha': 0.8}

labels = ['Math', 'English', 'Physics', 'Programming']

fig, ax = plt.subplots(figsize=(10, 7))
ax.boxplot(data, labels=labels, medianprops=medianprops, boxprops=boxprops, whiskerprops=whiskerprops, capprops=whiskerprops)

ax.set_ylim ([0, 100])

major_yticks = np.arange(0, 101, 20)
minor_yticks = np.arange(0, 101, 5)

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(labelsize=20)
ax.tick_params(axis ='x', labelrotation=10)

ax.grid(axis='y', linewidth=2)
ax.grid(axis='y', which='minor', linewidth=2, linestyle=':')
ax.grid(axis='x', linewidth=0)
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=15, size=(100, 1))
chem_scores = np.random.normal(loc=70, scale=10, size=(n_student, 1))
phy_scores = np.random.normal(loc=30, scale=12, size=(n_student, 1))
pro_scores = np.random.normal(loc=80, scale=5, size=(n_student, 1))

data = np.hstack((math_scores, chem_scores, phy_scores, pro_scores))

medianprops = {'linewidth': 1.5, 'color': 'r'}
boxprops = {'linewidth': 1.5, 'color': 'k', 'alpha': 0.7}
whiskerprops = {'linestyle': '--', 'color': 'b', 'alpha': 0.8}

labels = ['Math', 'English', 'Physics', 'Programming']

fig, axes = plt.subplots(2, 1, figsize=(15, 10))
axes[0].boxplot(data)
axes[1].violinplot(data)

axes[0].set_ylim([0, 100])
axes[1].set_ylim([0, 100])

axes[0].tick_params(labelsize=20, bottom=False, labelbottom=False)
axes[1].tick_params(labelsize=20)

fig.subplots_adjust(hspace=0.1)
```
