- Figure : matplotlib은 한 번에 한장의 그림을 그린다. 이 그림을 가리키는 용어가 figure이다. figure는 그림을 그리는 캔버스로 생각할 수 있으며, 하나의 figure에는 여러 개의 plot이 들어갈 수 있다. plot이 하나의 차트에 해당한다.
- Axes : 보통 plot으로 생각하는 하나의 그래프. 각각의 Axes는 개별적으로 제목 및 x/y 레이블을 가질 수 있다. (왜 플롯이나 차트라고 하지 않는지…)
- Axis : 번역하면 ‘축’인데, 그려지는 축을 말하는 것이 아니라 정확히는 x, y 의 제한 범위를 말한다.
- Artist : Figure 위에 그려지는 모든 것들은 각각 artist라 한다. 텍스트, 라인 같은 것들. 이들 대부분은 자신이 그려지는 Axes에 묶여 있다고 볼 수 있다. (대부분이라는 것은 일부는 또 아니라는 말이다…)

```python
import matplotlib.pyplot as plt
import numpy as np
```

## plt.figure : Making Figures

```python
fig = plt.figure()   # ax 없는 빈 figure 생성 (후에 ax 추가)

plt.show()
```

## fig.add_subplot : Adding subplots

```python
fig= plt.figure(figsize=(5,5), facecolor='linen')

ax = fig.add_subplot()
ax.plot([2, 3, 1])
ax.scatter([2, 3, 1], [2, 3, 4])

plt.show()
```

## fig.suptitle : Title of Figure

```python
figsize = (5, 5)
fig, ax = plt.subplots(figsize = figsize)
fig.suptitle("Title of a Figure", fontsize=15, fontfamily='monospace')
```

## ax.set_title: Titel of Ax

```python
figsize = (5, 5)
fig, ax = plt.subplots(figsize = figsize)
ax.set_title("Title of a Ax", fontsize=15, fontfamily='monospace')
```

## ax.set_xlabel and ax.set_ylabel : Usage

```python
figsize = (5, 5)
fig, ax = plt.subplots(figsize = figsize)

ax.set_xlabel("X label", fontsize=10)
ax.set_ylabel("Y label", fontsize=10)
```

## ax.set_xlabel and ax.set_ylabel : Arguments



```python
figsize = (5, 5)
fig, ax = plt.subplots(figsize = figsize)

ax.set_xlabel("X label", fontsize=10, color='darkblue', alpha=0.7)
ax.set_ylabel("Y label", fontsize=10, color='darkblue', alpha=0.7)
```

## ax.set_xlabel and ax.set_ylabel : with Title

```python
figsize = (5, 5)
fig, ax = plt.subplots(figsize = figsize)

fig.suptitle("Title of a Figure", fontsize=20, color='darkblue', alpha=0.9)

ax.set_xlabel("X label", fontsize=10, color='darkblue', alpha=0.7)
ax.set_ylabel("Y label", fontsize=10, color='darkblue', alpha=0.7)
```

## fig.tight_layout(ax.set_title, ax.set_xlabel, ax.set_ylabel)

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.set_title("Title!", fontsize=20)
ax.set_xlabel("X Label!", fontsize=15)
ax.set_ylabel("X Label!", fontsize=15)

fig.tight_layout()
```

## ax.twinx

```python
fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot()
ax2 = ax1.twinx()   # ax.twinx: 2개의 그래프를 하나의 공간에 생성

ax1.set_xlim([0, 100])    # set_xlim: x축의 범위를 설정
ax1.set_ylim([0, 100])    # set_ylim: y축의 범위를 설정
ax2.set_ylim([0, 0.1])

ax1.set_title("Twinx Graph", fontsize=30)
ax1.set_ylabel("Data1", fontsize=20)
ax2.set_ylabel("Data2", fontsize=20)

fig.tight_layout()
```

## ax.tick_params : Tick Locations

```python
figsize = (5, 5)
fig, ax = plt.subplots(figsize = figsize)

ax.tick_params(labelsize=20, length=10, width=3, bottom=False, labelbottom=False, top=True, labeltop=True)

ax.tick_params(labelsize=20, length=10, width=3, bottom=False, labelbottom=False, left=False, labelleft=False, top=True, labeltop=True, right=True, labelright=True)
```

## ax.tick_params : x, y Axis Ticks

```python
figsize = (5, 5)
fig, ax = plt.subplots(figsize = figsize)

ax.tick_params(labelsize=20, length=10, width=3, rotation=30)

ax.tick_params(axis='x', labelsize=20, length=10, width=3, rotation=30)

ax.tick_params(axis='y', labelsize=20, length=10, width=3, rotation=30)
```

##  ax.set_xticks : Regular Intervals


```python
fig, ax = plt.subplots(figsize=(14, 7))

xticks = [i for i in range(0, 101, 20)]
ax.set_xticks(xticks)

ax.tick_params(axis='x', labelsize=20, length=10, width=3, rotation=30)
```

## ax.set_xticks: Major and Minor Ticks

```python
fig, ax = plt.subplots(figsize=(14, 7))

major_xticks = [i for i in range(0, 101, 20)]
minor_xticks = [i for i in range(0, 101, 5)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)

ax.tick_params(axis='x', labelsize=20, length=10, width=3, rotation=30)
ax.tick_params(axis='x', which='minor', length=5, width=2)
```

## ax.set_xticks: x, y Ticks

```python
fig, ax = plt.subplots(figsize=(14, 7))

major_xticks = [i for i in range (0, 101, 20)]
minor_xticks = [i for i in range (0, 101, 5)]

major_yticks = [i for i in range (0, 11, 2)]
minor_yticks = [i for i in range (0, 11, 1)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(axis='x', labelsize=20, length=10, width=3, rotation=30)
ax.tick_params(axis='x', which='minor', length=5, width=2)

ax.tick_params(axis='y', labelsize=20, length=10, width=3)
ax.tick_params(axis='y', which='minor', length=5, width=2)
```
